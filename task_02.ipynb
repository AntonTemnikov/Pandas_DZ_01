```python
import pandas as pd
data = pd.read_csv('power.csv')

filtered_countries = data [ (( data['country'] == 'Estonia') | 
                           (data['country'] == 'Latvia') | 
                           (data['country'] == 'Lithuania')) &
                            
                           ((data['category'] == 4) |
                            (data['category'] == 12) |
                            (data['category'] == 21)) &
                                      
                           (( data['year'] >= 2005) & 
                            ( data['year'] <= 2010)) &
                          
                              ((data['quantity']>0))]

filtered_countries.groupby(['country','year', 'category'])[['quantity']].sum().reset_index()

```
